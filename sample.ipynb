```python
# Embed
from langchain.embeddings.openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
# Text splitter
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
splits = text_splitter.split_documents(documents)
```

```python
from langchain.vectorstores import Chroma

db = Chroma.from_documents(
    documents=splits,
    embedding=embeddings,
    persist_directory="./chroma_langchain_db",
)

db.persist()
```

```python
#Retrieveal tool
from langchain.agents import tool

from langchain.agents import tool

#Returns top 5 Simimlar chunks in "Document format, then format it to list of dictioranies"

@tool
def query_vector_db_with_page_info(query: str) -> list:
    """
    Query vector database and return results with text and page information.
    Args:
        query: Search query string
    Returns:
        List of dictionaries with text content and page numbers from relevant document splits
    """
    results = db.similarity_search(query, k=5)

    # Convert each Document to a simple dict
    formatted = []
    for doc in results:
        formatted.append({
            "text": doc.page_content,
            "page": doc.metadata.get("page")
        })

    return formatted

```

```python
# importing llm as a agent

from langchain.llms import OpenAI
from langchain.agents import initialize_agent, AgentType
llm = OpenAI(model="gpt-3.5-turbo-instruct", temperature=0.2)

agent = initialize_agent([query_vector_db_with_page_info], llm, agent=AgentType.STRUCTURED_CHAT_ZERO_SHOT_REACT_DESCRIPTION , verbose=True)
```

```python
# Querying
from langchain.llms import OpenAI
from langchain_core.prompts.prompt import PromptTemplate
from collections import 


template = """
You are a researcher analyzing expert from documents. Your task is to answer the question using only the provided information in the "context" section below. Respond in clear and simple Arabic suitable for a general audience.

Instructions:
- إذا ما لقيت إجابة كافية في السياق، قل بصراحة "المعلومة غير متوفرة في الوثيقة" 

Context: {context}
Question: {query}
Answer:
"""


query = "ماهي المادة الثالثة والتسعون "

returned_docs = query_vector_db_with_page_info(query) # has 5 similir dictioranies similar to the query above

def format_context(returned_docs):
    """
    function that converts list of dictionaries into one string
    """
    formatted_context = ""
    for doc in returned_docs:
        formatted_context  += f"{doc['text']} [page: {doc['page']}]" + "\n"
        
    return formatted_context

prompt_template = PromptTemplate.from_template(template)

context = format_context(returned_docs)

prompt = prompt_template.format(
    context=context,
    query=query,
)

refrences = []

for doc in returned_docs:
    refrences.append(doc['page']+1)


response_text = llm.invoke(prompt)




print(response_text)
print("-"*10)
print("راجع الصفحات حسب الترتيب: ")
for reference in refrences:
    print(reference)

```
